Import JHU Global Data

```
def import_JHU_Global():
  ## Basic Import:
  # Total Cases
  dfTC = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
  dfTC = dfTC.drop(['Lat','Long'],axis=1)
  dfTC = grouping = dfTC.groupby('Country/Region').sum()
  # Recoveries
  dfR = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
  dfR = dfR.drop(['Lat','Long'],axis=1)
  dfR = grouping = dfR.groupby('Country/Region').sum()
  # Fatalities
  dfD = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
  dfD = dfD.drop(['Lat','Long'],axis=1)
  dfD = grouping = dfD.groupby('Country/Region').sum()

  ## Basic Format:
  # Get Index & Header
  dfIndex = dfTC.iloc[0: ,0 ]
  dfHeader = dfTC.iloc[0 , 2:]

  # Make sure columns in dfTC, dfR and dfD are the same
  df = pd.merge(dfTC,dfR, on=['Country/Region'])
  df = pd.merge(df,dfD, on=['Country/Region'])
  s = int(round(len(df.columns)/3))

  dfTC = df.iloc[0:,0:s]
  dfR = df.iloc[0:,s:2*s]
  dfD = df.iloc[0:,2*s:]

  dfTC.columns = dfD.columns
  dfR.columns = dfD.columns

  # Invert to have countries as columns
  dfTC = dfTC.T
  dfR = dfR.T
  dfD = dfD.T

  # Give Date/index Column Header
  dfTC.index.name = 'Date'
  dfR.index.name = 'Date'
  dfD.index.name = 'Date'

  # Convert index to datetime format
  dfTC.index = pd.to_datetime(dfTC.index)
  dfR.index = pd.to_datetime(dfR.index)
  dfD.index = pd.to_datetime(dfD.index)

  # Name Columns: Country 
  dfTC.columns.name = 'Country'
  dfR.columns.name = 'Country'
  dfD.columns.name = 'Country'

  # Make sure Datatype is correct:
  df = df.apply(pd.to_numeric, errors='coerce')

  # Sort Columns to be in alphabetical order
  dfTC = dfTC.reindex(sorted(dfTC.columns), axis=1)
  dfR = dfR.reindex(sorted(dfR.columns), axis=1)
  dfD = dfD.reindex(sorted(dfD.columns), axis=1)

  return dfTC, dfR, dfD

```

```
def import_OxCGRT():
  # Import database
  dfOxCGRT = pd.read_csv("https://raw.githubusercontent.com/OxCGRT/covid-policy-tracker/master/data/OxCGRT_latest.csv")
  
  # Tabulate data
  dfSI = dfOxCGRT.pivot(index='Date', columns='CountryName', values='StringencyIndex')
  
  # Convert Index to Date Format
  dfSI.index = pd.to_datetime(dfSI.index, format="%Y%m%d")

  # Name Columns: Country 
  dfSI.columns.name = 'Country'

  return dfSI
```

Format OxCGRT to JHU Global Format

```
def format_dfSI_to_JHU_Global(dfSI,dfTC):
  # Removing OxCGRT columns not found in JHU
  list_SI_countries_remove = []
  for j in range(0,len(dfSI.columns)-1):
    if dfSI.columns.values[j] in dfTC.columns:
      continue
    else:
      # print("dfSI: ", dfSI.columns.values[j]) # Print SI Countries to be removed
      list_SI_countries_remove.append(dfSI.columns.values[j])

  # print("List of removed countries from SI: ", list_SI_countries_remove)
  dfSI = dfSI.drop(list_SI_countries_remove,axis=1)

  # Make SI start date the same as TC
  dfSI = dfSI.drop(dfSI.index[0:dfSI.index.get_loc(dfTC.index.values[0])])

  # Make empty array for new SI columns:
  empty_dfSI_cols = [0] * len(dfTC.index)

  # Make dfSI columns = dfTC
  for i in range(0,len(dfTC.columns)-1):
    if dfTC.columns.values[i] in dfSI.columns:
      continue
    else:
      # Add dfTC country column to dfSI full of NaN values
      dfSI[dfTC.columns.values[i]] = np.nan 

  # Sort Columns to be in alphabetical order
  dfSI = dfSI.reindex(sorted(dfSI.columns), axis=1)

  return dfSI
```

Calculate d(t), y(t), b(t) & rf(t)

```
def calculate_dybrf(TC,I,D,R):
  # Get Shape JHU Global tables:
  rows, cols = TC.shape

  # Create empty np.array for d(t), y(t), b(t) and Rf
  d = np.empty_like(TC)
  y = np.empty_like(TC)
  b = np.empty_like(TC)
  rf = np.empty_like(TC)

  # Determine and Fill parameters
  for i in range(1,rows):
      for j in range(0,cols):
          # d(t), y(t) & b(t)
          if I[i-1][j] > 0:
              d[i][j] = (D[i][j] - D[i-1][j]) / I[i-1][j]
              y[i][j] = (R[i][j] - R[i-1][j]) / I[i-1][j]
              b[i][j] = (I[i][j] - I[i-1][j] + 
                         R[i][j] - R[i-1][j] + 
                         D[i][j] - D[i-1][j]) / I[i-1][j]
          else:
              d[i][j] = 0.
              y[i][j] = 0.
              b[i][j] = 0.
          
          # rf(t)
          if (y[i][j] + d[i][j]) > 0:
              rf[i][j] = b[i][j] / (y[i][j] + d[i][j])
          else:
              rf[i][j] = 0.

  return d,y,b,rf
```

Calculate y_sg(t), d_sg(t), b_y_sg(t), rf_y_sg(t)

```
def calculate_params_y_sg(TC,I,R,D):
  # Get Shape JHU Global tables:
  rows, cols = TC.shape

  # Create empty np.array
  y_sg = np.empty_like(TC)
  d_sg = np.empty_like(TC)
  b_y_sg = np.empty_like(TC)
  rf_y_sg = np.empty_like(TC)

  # Calculate parameters using Savitzky-Golay
  for i in range(0,rows):
      for j in range(0,cols):

          # y_sg
          if i > 4 and i < (rows - 5) and I[i][j] > 0:
              y_sg[i][j] = (( (-4 * R[i-4][j]) + 
                              (-3 * R[i-3][j]) + 
                              (-2 * R[i-2][j]) + 
                              (-1 * R[i-1][j]) + 
                              (1  * R[i+1][j]) + 
                              (2  * R[i+2][j]) + 
                              (3  * R[i+3][j]) + 
                              (4  * R[i+4][j]) )/60) /I[i][j]

          elif i < 5 or i > (rows - 5):
              y_sg[i][j] = np.NaN
          else:
              y_sg[i][j] = 0.
          
          # d_sg
          if i > 4 and i < (rows - 5) and I[i][j] > 0:
              d_sg[i][j] = (( (-4 * D[i-4][j]) + 
                              (-3 * D[i-3][j]) + 
                              (-2 * D[i-2][j]) + 
                              (-1 * D[i-1][j]) + 
                              ( 1 * D[i+1][j]) + 
                              ( 2 * D[i+2][j]) + 
                              ( 3 * D[i+3][j]) + 
                              ( 4 * D[i+4][j]) )/60) /I[i][j]
          elif i < 5 or i > (rows - 5):
              d_sg[i][j] = np.NaN
          else:
              d_sg[i][j] = 0.

          # b(t)[y_sg]
          if  I[i-1][j] > 0 and i < (rows - 5):
              b_y_sg[i][j] = (I[i][j]/ I[i-1][j]) - 1 + d_sg[i][j] + y_sg[i][j]
          elif i < 5 or i > (rows - 5):
              b_y_sg[i][j] = np.NaN
          else:
              b_y_sg[i][j] = d_sg[i][j] + y_sg[i][j]

          # Rf(t)[y_sg]
          if (y_sg[i][j] + d_sg[i][j]) > 0 and i < (rows - 5):
              rf_y_sg[i][j] = b_y_sg[i][j] / (y_sg[i][j] + d_sg[i][j])
          elif i < 5 or i > (rows - 5):
              rf_y_sg[i][j] = np.NaN
          else:
              rf_y_sg[i][j] = 0.

  return y_sg, d_sg, b_y_sg, rf_y_sg
```

Calculate Error in d(t), y(t) & b(t)

```
def calculate_dyb_error(TC,I,R,D,d,y,b):
  # Get Shape JHU Global tables:
  rows, cols = TC.shape

  # Create empty np.array
  d_error = np.empty_like(TC)
  y_error = np.empty_like(TC)
  b_error = np.empty_like(TC)
  
  # Calculate d(t)_error , y(t)_error , b(t)_error
  for i in range(1,rows):
      for j in range(0,cols):
          # d(t)_error:
          if I[i-1][j] * (D[i][j]-D[i-1][j]) > 0:
              d_error[i][j] = d[i][j] * np.sqrt((1/(D[i][j]-D[i-1][j]))+(1/(I[i-1][j])))
          elif I[i-1][j] * (D[i][j]-D[i-1][j]) == 0:
              d_error[i][j] = d_error[i-1][j] 
          
          # y(t)_error:
          if I[i-1][j] * (R[i][j]-R[i-1][j]) > 0:
              y_error[i][j] = y[i][j] * np.sqrt((1/(R[i][j]-R[i-1][j]))+(1/(I[i-1][j])))
          elif I[i-1][j] * (R[i][j]-R[i-1][j]) == 0:
              y_error[i][j] = y_error[i-1][j] 
          
          # b(t)_error:
          if I[i-1][j] * (I[i][j]-I[i-1][j] + R[i][j]-R[i-1][j] + D[i][j]-D[i-1][j]) > 0:
              b_error[i][j] = b[i][j] * np.sqrt((1/(I[i][j]-I[i-1][j] + R[i][j]-R[i-1][j] + D[i][j]-D[i-1][j]))+(1/(I[i-1][j])))
          elif I[i-1][j] * (I[i][j]-I[i-1][j] + R[i][j]-R[i-1][j] + D[i][j]-D[i-1][j]) == 0:
              b_error[i][j] = b_error[i-1][j]

  return d_error, y_error, b_error
```

Build Time Series Global

```
def build_timeseries_global(dfTC, dfI, dfR, dfD, dfyt, dfyt_avg, dfdt, dfbt, dfrf, dfrf_yavg, dfSI):

  df_timeseries_global = pd.merge(dfTC.stack().reset_index(),dfI.stack().reset_index(), on=['Date','Country'])             # TC, I
  df_timeseries_global = pd.merge(df_timeseries_global,dfR.stack().reset_index(), on=['Date','Country'])                   # R
  df_timeseries_global = pd.merge(df_timeseries_global,dfD.stack().reset_index(), on=['Date','Country'])                   # D
  df_timeseries_global = pd.merge(df_timeseries_global,dfyt.stack().reset_index(), on=['Date','Country'])                  # yt

  df_timeseries_global = pd.merge(df_timeseries_global,dfyt_avg.stack().reset_index(), on=['Date','Country'])              # y_avg

  df_timeseries_global = pd.merge(df_timeseries_global,dfdt.stack().reset_index(), on=['Date','Country'])                  # dt
  df_timeseries_global = pd.merge(df_timeseries_global,dfbt.stack().reset_index(), on=['Date','Country'])                  # bt   
  df_timeseries_global = pd.merge(df_timeseries_global,dfrf.stack().reset_index(), on=['Date','Country'])                  # rf 

  df_timeseries_global = pd.merge(df_timeseries_global,dfrf_yavg.stack().reset_index(), on=['Date','Country'])            # rf (y_avg) 

  df_timeseries_global = pd.merge(df_timeseries_global,dfSI.stack().reset_index(), on=['Date','Country'])            # rf (y_avg) 

  df_timeseries_global.columns = ['Date', 'Country', 'TC', 'I', 'R', 'D', 'yt', 'yt_avg', 'dt', 'bt', 'rf', 'rf_yavg','p']

  return df_timeseries_global
```

```
def calculate_dybrf(dataSeries):
  frames = []
  for k in range(len(dataSeries['NAME'].unique())):
      print(dataSeries['NAME'].unique()[k])
      data = dataSeries[dataSeries['NAME']==dataSeries['NAME'].unique()[k]]
      data = data.reset_index()
      data =  data.drop(['index'], axis=1)
      i = 1
      while(i < data.shape[0]):
          # d(t), y(t) & b(t)
          if data['I'][i-1] > 0:
              data.at[i, 'd'] = (data['D'][i] - data['D'][i-1]) / data['I'][i-1]
              data.at[i, 'y'] = (data['R'][i] - data['R'][i-1]) / data['I'][i-1]
              data.at[i, 'b'] = ((data['I'][i] - data['I'][i-1]) + 
                         (data['R'][i] - data['R'][i-1]) + 
                         (data['D'][i] - data['D'][i-1])) / data['I'][i-1]
          else:
              data.at[i, 'd'] = 0.
              data.at[i, 'y'] = 0.
              data.at[i, 'b'] = 0.
          
          # rf(t)
          if (data['y'][i] + data['d'][i]) > 0:
              data.at[i, 'rf'] = data['b'][i]/ (data['y'][i] + data['d'][i])
          else:
              data.at[i, 'rf'] = 0.

          dataD = data['D'][i] - data['D'][i-1]
          dataR = data['R'][i] - data['R'][i-1]
          dataI = data['I'][i] - data['I'][i-1]
          tot = dataD+dataR+dataI
          
          # if data['I'][i-1] * (dataI + dataR + dataD) > 0:
          #     beta_error = data['b'][i] * math. sqrt((1/(tot)) + (1/data['I'][i-1]))
          #     # print(tot,data['I'][i-1],beta_error,data['bt'][i])
          #     data.at[i, 'bt error'] = beta_error
              
          
          # elif data['I'][i-1] * (dataI + dataR + dataD) == 0:
          #     data.at[i, 'bt error'] = data['bt error'][i-1]

          # b(t)_error:
          if data['I'][i-1] * (data['I'][i]-data['I'][i-1] + data['R'][i]-data['R'][i-1] + data['D'][i]-data['D'][i-1]) > 0:
              data.at[i, 'bt error'] = data['b'][i] * np.sqrt((1/(data['I'][i]-data['I'][i-1] + data['R'][i]-data['R'][i-1] + data['D'][i]-data['D'][i-1]))+(1/(data['I'][i-1])))
          elif data['I'][i-1] * (data['I'][i]-data['I'][i-1] + data['R'][i]-data['R'][i-1] + data['D'][i]-data['D'][i-1]) == 0:
              # data.at[i, 'bt error'] = data['bt error'][i-1]
              data.at[i, 'bt error'] = 1
          
          i += 1
      frames.append(data)
  CompleteData = pd.concat(frames)
  return CompleteData
```
